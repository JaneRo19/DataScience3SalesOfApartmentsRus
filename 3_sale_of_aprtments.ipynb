**Обработка неявных дубликатов в столбце locality_name**

Выведем еще раз названия населенных пунктов для того, чтобы определить есть ли какие-то скрытые дубликаты в данных, наример, названия одного и того же населенного пунка по-разному. Если данную ошибку не поправить, то при группировке данных можно получить неверные значения.

Выведем на экран названия населенных пунктов и отсортируем их в порядке убывания.

```python
# отсортируем названия населенных пунктов и выведем на экран
data_locality_name = data['locality_name']
data_locality_name_sorted = data_locality_name.sort_values()
data_locality_name_sorted = data_locality_name_sorted.unique()
print(data_locality_name_sorted)
```

## Расчёты и добавление результатов в таблицу

Рассчитаем и добавим в таблицу следующие значения:

    - цену квадратного метра (cost_per_square_meter)
    - день недели публикации объявления (weekday_first_exposition)
    - месяц публикации объявления (month_first_exposition)
    - год публикации объявления (year_first_exposition)
    - этаж квартиры (первый, последний, другой)
    - соотношение жилой и общей площади (ratio_living_total_area)

```python
# добавим столбец стоимость квадратного метра
data['cost_per_square_meter'] = (data['last_price'] / data['total_area']).round(2)
```

```python
# добавим столбец день недели публикации объявления
data['weekly_first_exposition'] = data['first_day_exposition'].dt.weekday
```

```python
# добавим столбец месяц публикации объявления
data['month_first_exposition'] = data['first_day_exposition'].dt.month
```

```python
# добавим столбец год публикации объявления
data['year_first_exposition'] = data['first_day_exposition'].dt.year
```

```python
# добавим столбец отношения жилой и общей площади
data['ratio_living_total_area'] = (data['living_area'] / data['total_area']).round(2)
```

```python
# добавим столбец отношения площади кухни и общей площади
data['ratio_kitchen_total_area'] = (data['kitchen_area'] / data['total_area']).round(2)
```

```python
# напишем функцию для получения названия этажа квартиры
def get_appartment_floor(data):
    """
    Возвращает названия этажа : первый, последний, другой
    
    param: data - датафрейм
    """
    if (data['floor'] == 1)|((data['floors_total'] == 1)&(data['floor'] == data['floors_total'])):
        return 'первый'
    elif data['floor'] == data['floors_total']:
        return 'последний'
    return 'другой'
```

```python
# добавим столбец с названием этажа
data['appartment_floor'] = data.apply(get_appartment_floor, axis=1)
```

**Зависимость стоимости квадратного метра от дня недели размещения объявления**

```python
# для наглядности графика создадим дополнительный столбец с названием дней недели
def weekly_day(data):
    """
    Возвращает название дня недели
    
    """    
    list_numbers = [0, 1, 2, 3, 4, 5, 6]
    list_day_names = ['1_понедельник', '2_вторник', '3_среда', '4_четверг', '5_пятница', '6_суббота', '7_воскресенье']
    
    for i in range(len(list_numbers)):
        if data['weekly_first_exposition'] == list_numbers[i]:
            return list_day_names[i]    
```

```python
data['weekly_day_name'] = data.apply(weekly_day, axis=1)
```

```python
# создадим сводную таблицу
pivot_table_weekly_first_exposition = (data.pivot_table(index='weekly_day_name', values='cost_per_square_meter', aggfunc='median')
                                       .plot(grid=True, figsize=(8, 4))
                                      )
plt.xlabel('День недели')
plt.ylabel('Стоимость квадратного метра')
pivot_table_weekly_first_exposition = (data.pivot_table(index='weekly_day_name', values='last_price_thouthands', aggfunc='median')
                                       .plot(grid=True, figsize=(8, 4))
                                      )
plt.xlabel('День недели')
plt.ylabel('Стоимость кваритиры в тысячах')
plt.show()
```

На приведенном графике сильной зависимоти стоимости квадратного метра от дня недели размещения объявления нет, в целом максимум и минимум отличается на 2000. Но можно заметить, что самые высокие цены в понедельник, возможно, это связано с тем, что люди в начале недели выкладывают объявление и завышают цены, а позже под конец недели эти цены немного падают.

### Определим 10 населенных пунктов с наибольшим числом объявлений

Определим 10 населенных пунктов с наибольшим числом объявлений о продаже, исключив Санкт-Петербург, так как мы его будем рассматривать отдельно. Определим их среднюю цену квадратного метра. И выберем среди них населенные пункты с высокой и низкой ценой.

Для того, чтобы получить однородные данные для подсчета, создадим столбец статус города, в котором будем, указывать Санкт-Петербург это или пригород. Создадим фунцию.

```python
def town_status(data):
    """
    Возвращает статус города
    """
    if data['locality_name'] == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    else:
        return 'пригород'
```

```python
# применим функцию
data['town_status'] = data.apply(town_status, axis=1)
```

Проведем груупировку сначала по количеству объявлений в каждом населенном пункте, отберем 11 из них с наибольшим количеством объявлений после чего сделаем группировку в порядке убывания стоимости квадратного метра.

```python
data_top_10 = (data.groupby('locality_name')
               .agg({'town_status':'count', 'cost_per_square_meter':'mean'})
               .sort_values('town_status', ascending=False)
               .head(11)
               .sort_values('cost_per_square_meter', ascending=False)
               .round({'cost_per_square_meter':2})
              )
```

```python
display(data_top_10)
```

Таким образом, получили, что в пригороде:

    - наибольшая стоимость жилья за квадратный метр в городе Пушкин: 103125,82
    - наименьшая стоимсоть жилья за квадратный метр в городе Выборг: 58141,91
